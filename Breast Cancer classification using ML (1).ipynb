```python
import time
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import plot_tree
from matplotlib.backends.backend_pdf import PdfPages

# Create and train the Random Forest classifier
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train_scaled, y_train)

# Save the feature importances
feature_importances = random_forest.feature_importances_
feature_importance_dict = dict(zip(X_train.columns, feature_importances))
sorted_features = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)

# Plot and save the trees as PDF files
with PdfPages('random_forest_trees.pdf') as pdf:
    # Plot each decision tree in the Random Forest
    for i, tree in enumerate(random_forest.estimators_):
        start_time = time.time()
        plt.figure(figsize=(20, 10))
        plot_tree(tree, feature_names=X_train.columns, class_names=['benign', 'malignant'],
                  filled=True, rounded=True)
        plt.title(f"Random Forest Tree {i+1}")
        pdf.savefig()
        plt.close()  # Close the figure to avoid displaying it
        end_time = time.time()
        print(f"Tree {i+1} - Time taken: {end_time - start_time:.6f} seconds")
```
